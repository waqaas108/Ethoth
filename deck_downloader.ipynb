```python
# Pretty much ChatGPT's code, but with my values
# Path to the images folder
folder_path = 'images/'
destination = 'deck/'

# Iterate over each image file in the folder
for filename in os.listdir(folder_path):
    if filename.endswith('.jpg') | filename.endswith('.webp') | filename.endswith('.png'):
        # Open the image
        image_path = os.path.join(folder_path, filename)
        destination_path = os.path.join(destination, filename)
        image = Image.open(image_path)
        
        # Calculate the new height based on the aspect ratio and width
        width, height = image.size

        pros_width = int(height / 22 * 15)
        pros_height = int(width / 15 * 22)

        if pros_width > width:
            new_height = pros_height

            # Calculate the top and bottom coordinates for cropping
            top = (height - new_height) // 2
            bottom = top + new_height
            
            # Crop the image
            cropped_image = image.crop((0, top, width, bottom))
            
            # Resize the image to a width of 360 pixels
            resized_image = cropped_image.resize((360, 528))
            
            # Save the edited image
            resized_image.save(destination_path)

        else:
            new_width = pros_width

            # Calculate the left and right coordinates for cropping
            left = (width - new_width) // 2
            right = left + new_width

            # Crop the image
            cropped_image = image.crop((left, 0, right, height))

            # Resize the image to a width of 360 pixels
            resized_image = cropped_image.resize((360, 528))

            # Save the edited image
            resized_image.save(destination_path)
```

```python
# preliminary attempt at creating a deck class, ultimately useful for name and image retrieval and editing
import os
import random
from .cards.card import Card
import pandas as pd

cards_df = pd.read_csv("cards.csv")

deck = []
for index, row in cards_df.iterrows():
    card = Card(row["Card"], os.path.join("deck", row["Image"], row["Title"], row["planet_orb"], row["planet_house"], row["sign_1"], row["sign_2"], row["sign_3"], row["suit_1"], row["suit_2"], row["path"], row["sephira"], row["element_1"], row["element_2"]))
    deck.append(card)



class Deck:
    def __init__(self, images_folder="deck"):
        self.cards = []

        for filename in os.listdir(images_folder):
            if filename.endswith(".jpg") or filename.endswith(".jpeg"):
                card_title = os.path.splitext(filename)[0]
                card_title = " ".join([word.capitalize() if word != "of" else word for word in card_title.split("_")])
                self.cards.append(Card(card_title, os.path.join(images_folder, filename)))

    def shuffle(self):
        random.shuffle(self.cards)
```
